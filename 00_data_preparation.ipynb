## 1. What is Movie Recommender System?
`Definition`: A movie recommender system is a computer system that suggests movies to users based on their preferences, past behavior, or similarities with other users.


Types of Recommender System
1. `Content based`: depends on similarities of content, example when we watch video on sports only sports related videos are suggested.
2. `Collaboration filtering based`: depends on users interests, example when user have similar interest the contets are recommended between each other.
3. `Hybrid based`: uses both content and collaboration based recommender system.

#### Bringing the columns into correct format

the movie dataframe must be converted to new dataframe with only three important columns
movie-id | title | tags

tags are the concatenation of all the "genres", "keywords", "title", "overview", "cast", "crew" columns

```python
import ast

def listToWords(obj):
    final_words = []
    for dic in ast.literal_eval(obj):
        if dic.get("name", None):
            final_words.append(dic["name"])
    return final_words
```

```python
movies["genres"] = movies["genres"].apply(listToWords)
```

```python
movies.head()
```

```python
movies["keywords"] = movies["keywords"].apply(listToWords)
```

```python
movies.head()
```

```python
def listToWords1(obj):
    final_words = []
    counter = 0
    for dic in ast.literal_eval(obj):
        if dic.get("name", None) and counter != 3:
            final_words.append(dic["name"])
            counter += 1
        else:
            break
    return final_words
```

```python
movies["cast"] = movies["cast"].apply(listToWords1)
```

```python
movies.head()
```

```python
def listToWords2(obj):
    final_words = []
    for dic in ast.literal_eval(obj):
        if dic["job"] == "Director":
            final_words.append(dic["name"])
            break
    return final_words
```

```python
movies["crew"] = movies["crew"].apply(listToWords2)
```

```python
movies.head()
```

```python
movies["overview"] = movies["overview"].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
movies["genres"]= movies["genres"].apply(lambda x: [i.replace(" ", "") for i in x])
movies["keywords"]= movies["keywords"].apply(lambda x: [i.replace(" ", "") for i in x])
movies["cast"]= movies["cast"].apply(lambda x: [i.replace(" ", "") for i in x])
movies["crew"]= movies["crew"].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies["tags"] = movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
```

```python
movies.head()
```

```python
new_df = movies[["id", "title", "tags"]]
new_df.head()
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x: x.lower())
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df["tags"] = new_df["tags"].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
vectors = cv.fit_transform(new_df["tags"]).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie, top_n=5):
    movie_index = new_df[new_df["title"] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:top_n+1]

    results = []
    for idx, score in movie_list:
        results.append({
            "id": int(new_df.iloc[idx].id),
            "title": new_df.iloc[idx].title,
            "score": float(score),
        })
    return results
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
